```python
import pandas as pd
import numpy as np
import os

# pip install chinesecalendar
from chinese_calendar import is_holiday

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.DataFrame(pd.date_range('2012-09-27 00:00:00', '2012-10-03 00:00:00', freq='12H'))
df.columns = ['time']
```

```python
df
```

```python
def get_time_feature(df, col):
    
    prefix = col + "_"
    
    df[col] = pd.to_datetime(df[col])
    df[prefix + 'year'] = df[col].dt.year
    df[prefix + 'month'] = df[col].dt.month
    df[prefix + 'day'] = df[col].dt.day
    df[prefix + 'hour'] = df[col].dt.hour
    df[prefix + 'weekofyear'] = df[col].dt.weekofyear
    df[prefix + 'dayofweek'] = df[col].dt.dayofweek
    df[prefix + 'is_wknd'] = df[col].dt.dayofweek // 5
    df[prefix + 'quarter'] = df[col].dt.quarter
    df[prefix + 'is_month_start'] = df[col].dt.is_month_start.astype(int)
    df[prefix + 'is_month_end'] = df[col].dt.is_month_end.astype(int)
    df[prefix + 'is_holiday'] = df[col].apply(lambda x: 1 if is_holiday(x) else 0)
    
    return df
```

```python
df = get_time_feature(df, "time")
```
